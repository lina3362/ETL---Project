1, What is the average salary with benefits by year? Show it in a dataframe and a chart

```python
avg_salary_by_year = pd.read_sql('''SELECT Year, ROUND(AVG(TotalPayBenefits), 2) AS Avg_Total_Pay
FROM salaries
GROUP BY Year
ORDER BY Year DESC;
''', conn)
avg_salary_by_year
```

```python
# Create a line chart of average salary by title.
fig, ax = plt.subplots(figsize=(6,4))
fig = avg_salary_by_year.plot(kind="line", x="year", y="avg_total_pay", legend=False, rot=90,ax=ax)

ax.xaxis.set_ticks([2011,2012, 2013, 2014])
plt.title("Average Total Pay 2011 - 2014", fontsize=10)
plt.xlabel("Year")
plt.ylabel("Average Total Pay")
plt.tight_layout()

#save the figure as a png file
plt.savefig('../Output/Average Total Pay 2011 - 2014')

plt.show()
```

3, What are the top 10 job titles that have the highest salary?

```python
salary_top10_bytitle = pd.read_sql('''SELECT JobTitle, ROUND(AVG(TotalPayBenefits), 2) AS TotalPay
FROM salaries
GROUP BY JobTitle
ORDER BY TotalPay DESC
LIMIT 10;
''', conn)
salary_top10_bytitle
```

```python
salary_top10_bytitle.describe().round(2)
```

```python
# Create a bar chart of average salary by title.
fig, ax = plt.subplots(figsize=(10,10))
fig = salary_top10_bytitle.plot(kind="bar", x="jobtitle", y="totalpay", legend=False, rot=90,ax=ax)

plt.title("Top 10 Salary by Title", fontsize=10)
plt.xlabel("Title")
plt.ylabel("Salary")
plt.grid()
plt.tight_layout()

#save the figure as a png file
plt.savefig('../Output/Top 10 Salary by Title')

plt.show()
```

4, which 10 employees have the highest average pay in SF?

```python
salary_top10_byemployee = pd.read_sql('''SELECT EmployeeName, ROUND(AVG(TotalPayBenefits), 2) AS TotalPay
FROM salaries
GROUP BY EmployeeName
ORDER BY TotalPay DESC
LIMIT 10;
''', conn)
salary_top10_byemployee
```

```python
salary_top10_byemployee.describe().round(2)
```

```python
# Create a bar chart of average salary by title.
fig, ax = plt.subplots(figsize=(10,10))
fig = salary_top10_byemployee.plot(kind="bar", x="employeename", y="totalpay", legend=False, rot=90,ax=ax)

plt.title("Top 10 Salary by Employees", fontsize=10)
plt.xlabel("Employees")
plt.ylabel("Salary")
plt.grid()
plt.tight_layout()

#save the figure as a png file
plt.savefig('../Output/Top 10 Salary by Employees')

plt.show()
```

5, Top 10 employees who switched jobs the most in 4 years

```python
employee_job_count = pd.read_sql('''SELECT EmployeeName, COUNT(JobTitle) AS Job_Count, COUNT(Year) AS Year_Count, ROUND(AVG(TotalPayBenefits), 2) AS Avg_Total_Pay
FROM salaries
GROUP BY EmployeeName
ORDER BY Job_Count DESC
LIMIT 10;
''', conn)
employee_job_count
```

6, What are the average pay for fulltime employees vs partime employees?

```python
fulltime_parttime = pd.read_sql('''SELECT Status, ROUND(AVG(TotalPayBenefits), 2) AS Avg_Total_Pay
FROM salaries
WHERE Status IN ('FT','PT')
GROUP BY Status
ORDER BY Avg_Total_Pay DESC;
''', conn)
fulltime_parttime
```

7, Top 10 most needed jobs in City of San Francisco

```python
Job_Title_counts = pd.read_sql('''SELECT JobTitle, COUNT(JobTitle) AS Job_Count
FROM salaries
GROUP BY JobTitle
ORDER BY Job_Count DESC
LIMIT 10;;
''', conn)
Job_Title_counts
```

```python
# Create a bar chart of average salary by title.
fig, ax = plt.subplots(figsize=(10,10))
fig = Job_Title_counts.plot(kind="bar", x="jobtitle", y="job_count", legend=False, rot=90,ax=ax)

plt.title("Top 10 Most Needed Job in SF", fontsize=10)
plt.xlabel("Job Title")
plt.ylabel("Counts")
plt.grid()
plt.tight_layout()

#save the figure as a png file
plt.savefig('../Output/Top 10 Most Needed Job in SF')

plt.show()
```
