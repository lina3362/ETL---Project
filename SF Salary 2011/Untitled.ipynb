```python
import psycopg2
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from config import username, password, database

from sqlalchemy import create_engine, inspect
```

```python
# Create Database Connection
engine = create_engine(f"postgresql://{username}:{password}@localhost:5432/{database}")
conn = engine.connect()
```

```python
# Create the inspector and connect it to the engine
inspector = inspect(engine)
inspector.get_table_names()
```

```python
#Create a dataframe shows salaries
Salary = pd.read_sql("SELECT * FROM salaries ORDER BY id", conn)

Salary.head()
```

```python
employee_2011_rich = pd.read_sql(''' SELECT EmployeeName, jobtitle, totalpaybenefits, Year  
                                FROM Salaries
                                WHERE Year = '2011' AND totalpaybenefits >= '100000'
                                       ''', conn)
employee_2011_rich
```

```python
trend = pd.read_sql("SELECT jobtitle, COUNT(jobtitle) AS Job_Count FROM Salaries GROUP BY jobtitle ORDER BY Job Count DESC", conn)
print(trend)
trend.plot(x="Year",y="Job_Count",legend=False,marker="o")
plt.title("Count of reported jobs every year")
plt.xticks(trend.Year)
plt.ylabel("No of jobs reported");
```
